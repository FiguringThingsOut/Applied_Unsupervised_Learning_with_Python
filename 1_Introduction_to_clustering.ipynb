### No Math K-Means Walkthrough
1. Pick k centeroid (k = expected distinct cluster)
2. Randomply place k centeroids anywher amongst your existing data
3. Calculate Euclidian Distance from each centeroid to all point in training data
4. Training data grouped with their nearest centeroid
5. calculate mean data point, change centeroid to that location
6. step 3 through - 5 again till the membership in each group no longer changes

**Just so you know, a non vectorized implementation of Euclidian distance is as follows :**

```python
def dist(a, b):
    return math.sqrt(math.pow(a[0] - b[0],2) + math.pow(a[1] - b[1],2))
```

**Step 1 : Now, implement the optimized Euclidean distance:**

```python
from scipy.spatial.distance import cdist
```

**Step 2 : Store the value of X**

```python
X[105:110]
```

**Step 3 : Calculate the distances and choose the index of the shortest distance as a cluster**

```python
for x in X[105:110] :
    calcs = []
    for c in centroids :
        calcs.append(dist(x, c))
    print(calcs, "cluster membership :", np.argmin(calcs, axis = 0))
```

**Step 4 : Define the k_means function as follows and initialize k-centroids randomly. Repeat
the process until the difference between new/old centroids equal 0 using the
while loop**

```python
def k_means(X, K):
# Keep track of history so you can see k-means in action
    # empty list to contain temporary result or perma result
    centroids_history = []
    labels_history = []
    
    # X.shape[0] equals to number of data in X
    # Basically using np.random.choice we want to get K number of random number from 0 - 1500.
    # those number will be used as index.
    # X[index] will be used as centroids
    rand_index = np.random.choice(X.shape[0], K)
    
    # we get 3 index from previous code. 
    # assign those index as centroids
    centroids = X[rand_index]
    
    # tract centroids record
    centroids_history.append(centroids)
    
    # adding line 24 and 26. because there's a miss between number of centroids_history & labels_history.
    # before adding this 2 line, number of labels_history = number of centroids_history -1
    # by adding that 2 lines of codes, I fix that irregularities
    # line 24 and 26 equals to code line 32 and 34
    labels = np.argmin(cdist(X, centroids), axis=1)

    labels_history.append(labels)
        
    while True:
        # Euclidean distances are calculated for each point relative to
        # centroids, and then np.argmin returns the index location of the
        # minimal distance - which cluster a point is assigned to
        labels = np.argmin(cdist(X, centroids), axis=1)
               
        labels_history.append(labels)
        
        # Take mean of points within clusters to find new centroids
        # take X from certain cluster and calculate its center.
        new_centroids = np.array([X[labels == i].mean(axis = 0) for i in range(K)])
        
        # track centroids record
        centroids_history.append(new_centroids)
            
        # If old centroids and new centroids no longer change, k-means is
        # complete and end. Otherwise continue
        if np.all(centroids == new_centroids):
            break
        centroids = new_centroids
    return centroids, labels, centroids_history, labels_history

# run the k_means(X, 3)
centers, labels, centers_hist, labels_hist = k_means(X, 3)
```
